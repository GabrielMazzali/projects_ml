## Simple Monte Carlo Simulation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
annual_contrib = 6000
mean_return = 0.07 # Annual mean return (as a decimal)
std_return = 0.15 # Annual standard deviation of returns
years = 40

# print(normal_deviation.mean())

final_value = []

for idx in range(10000):
    year_value = 0
    for i in range(years):
        normal_deviation = np.random.normal(mean_return, std_return,1)
        year_value = (year_value + annual_contrib) * (1 + normal_deviation)
    final_value.append(float(year_value))

print(f'Mean: {round(np.mean(final_value))}')
print(f'Median: {round(np.median(final_value))}')

plt.hist(final_value, bins=50, edgecolor='k', alpha=0.7, rwidth=0.85)
plt.xlabel('Final Investment Value MM')
plt.ylabel('Frequency')
plt.title('Monte Carlo Simulation for S&P 500 Investment')

plt.show()
```
